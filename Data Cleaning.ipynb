```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
# drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
# Drop columns which are not use for analysis
df.drop(columns="Not_Useful_Column", inplace=True)
```

```python
df
```

```python
# remove unwanted characters from Last name
df['Last_Name'] = df['Last_Name'].str.lstrip('.').str.lstrip('/').str.rstrip('_').str.strip()
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('Na', np.nan)
```

```python
# Standardize the Phone number format
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x) if pd.notnull(x) else x)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3]+'-'+x[3:6]+'-'+x[6:10] if pd.notnull(x) else x)
```

```python
df['Phone_Number']
```

```python
# Split the Address column
df['Street_Address'], df['City'], df['ZIP_Code'] = df['Address'].str.split(', ', 2).str
```

```python
df
```

```python
# replace N with No and Y with Yes
df['Paying Customer'] = df['Paying Customer'].replace({'N': 'No', 'Y': 'Yes'})
```

```python
# replace N with No and Y with Yes
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'N': 'No', 'Y': 'Yes'})
```
